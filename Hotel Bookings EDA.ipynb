```python
# Define functions for a quick summary

# Categorical features
def desc_obj(col):
    print('*** ' + col + ' ***')
    print(df[col].describe())
    print()
    print(df[col].value_counts())
    print()

# Numerical features
def desc_num(col):
    print('*** ' + col + ' ***')
    print(df[col].describe())
    print()
    df[col].plot(kind = 'hist')
    plt.show()
```

### Focus on city hotels

#### EDA

```python
# Subset city hotels
city = df[df['hotel'] == 'City Hotel'].copy()
```

```python
# Function for visualising the relationship between categorical features and target variable
# First subplot shows the total number of canceled and not canceled reservations for each
# unique value of the categorical variable. Second subplot shows the percentage of canceled 
# reservations for each unique value of the categorical variable.
def visual_cancel(col, rotate = True):
    
    fig, axes = plt.subplots(1,2,figsize=(16,8))
    sns.countplot(x = col, data = city, palette="Set2",ax = axes[0], hue = 'is_canceled')
    axes[0].set_title('Booking cancellations per ' + col)
    if rotate:
        axes[0].tick_params(labelrotation=65)
    
    cancel_dict = {}
    for value in city[col].unique():
        try:
            cancels = city[city[col] == value]['is_canceled'].value_counts()[1]
        except:
            cancels = 0
        bookings = city[city[col] == value]['is_canceled'].value_counts().sum()
        cancel_percentage = round(cancels / bookings, 2)
        cancel_dict[value] = cancel_percentage

    canc_percent = pd.DataFrame({'values': list(cancel_dict.keys()), 
                                 'cancel_percentage':list(cancel_dict.values())})

    sns.barplot(x = 'values', y = 'cancel_percentage', data = canc_percent, 
                palette="Set2",ax = axes[1])
    axes[1].set_title('Percentage of cancelled bookings per ' + col)
    if rotate:
        axes[1].tick_params(labelrotation=65)
```

##### Arrival date day of month

```python
# Aggregate day of month to week of month

def day_to_week(x):
    if x <= 7:
        return '1st week'
    elif x <= 14:
        return '2nd week'
    elif x <= 21:
        return '3rd week'
    else:
        return '4th week'
    
city['arrival_date_week_of_month'] = city['arrival_date_day_of_month'].map(day_to_week)
city.drop('arrival_date_day_of_month', axis = 1, inplace = True)
```

```python
visual_cancel('arrival_date_week_of_month', rotate = False)
```

##### Total number of days ->  Stays in week nights + Stays in weekend nights

```python
# Sum week and weekend nights to get total number of nights of stay

city['days_of_stay'] = city['stays_in_week_nights'] + city['stays_in_weekend_nights']
city['days_of_stay'].value_counts()
```

```python
# Keep only values that have a frequency higher than 40

day_indices = []

for value in city['days_of_stay'].unique():
    if (city['days_of_stay'].value_counts()[value] < 40) | (value == 0):
        day_indices += list(city[city['days_of_stay'] == value].index)

city.drop(day_indices, inplace = True)
city.reset_index(drop = True, inplace = True)
city['days_of_stay'].value_counts()
```

##### Previous booking cancellation percentage 

```python
# Find percentage of cancelations for previous bookings, divide 'previous_cancellations' by sum 
# 'previous_cancellations' & 'previous_bookings_not_canceled'
def book(x):
    if x[0]+x[1] != 0:
        return round(x[0]/(x[0]+x[1]), 2)
    else: return 0
    
city['previous_bookings_cancel_percentage'] = city[['previous_cancellations', 'previous_bookings_not_canceled']].apply(book, axis = 1)
```

```python
# Check boxplot only for reservations that had past cancellations
fig, axes = plt.subplots(figsize=(8,5))
sns.boxplot(y = 'previous_bookings_cancel_percentage', x = 'is_canceled', 
            data = city[city['previous_bookings_cancel_percentage'] != 0], 
            palette="Set1", ax = axes)
axes.set_title('previous bookings cancel percentage (for booking with previous cancellations) vs is_canceled')
```

```python
# From the boxplot it is observed that the majority of bookings who had a percentage of previous
# cancelations up ~40% were not canceled this time (0), whereas most of the canceled bookings
# had 100% previous cancellations
```
